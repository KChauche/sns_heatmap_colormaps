# Seaborn Complete Colormap Demonstration 
This project aims to provide a demonstration of what each colormap available in Seaborn figures manifests as. This will allow others who work in data visualization to customize their figures to a greater degree. 

A write up of this project, along with information about colorblindness and which colormaps are compatible with people who have colorblindness is available at this blog: LINK 

```python
# import modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
# generates figures in jupyter notebook 
%matplotlib inline 
```

```python
# generates random set of numbers to use as data
df = np.random.rand(8, 10)
df = pd.DataFrame(df)
```

```python
# set mask so only lower half of heatmap is displayed
mask = np.triu(df.corr())

# generate a heatmap for each colormap name and export resulting figure
for color in colormaps: 
    plt.figure(figsize=(5, 5))
    sns.heatmap(df.corr(), mask=mask, cmap=color)
    plt.title(color);
    plt.savefig(f"./figures/{color}.png")
```
